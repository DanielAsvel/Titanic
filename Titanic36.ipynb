```python
import pandas as pd
import numpy  as np

from sklearn.preprocessing          import OneHotEncoder
from sklearn.neighbors              import KNeighborsClassifier # classificador por vizinhasa
from sklearn.model_selection        import train_test_split # modelo treino e teste
from sklearn.metrics                import accuracy_score # medir precisão do treino e teste
from sklearn.preprocessing          import MinMaxScaler # pré processamento dos dados. os valores iram ficar entre 0 ~ 1 
```

```python
base = pd.read_csv('https://raw.githubusercontent.com/DanielAsvel/Titanic/main/train.csv')
```

```python
base.head()
```

```python
base1 = base[['Sex',	'Age', 'Pclass']] # seprando as colunas que eu quero
```

```python
base1
```

```python
sex = base1.select_dtypes(include='object') # construindo uma coluna com só a sexo

onehot = OneHotEncoder(sparse=False, drop='first') # transformando sexo em binario
x_bin = onehot.fit_transform(sex)
x_bin
```

```python
num = base1.select_dtypes(exclude='object') # construindo uma variavel só com numoros

min = MinMaxScaler()  # transformando os valores para valores entre 0 e 1
x_num = min.fit_transform(num)
x_num
```

```python
X = np.append(x_bin, x_num, axis = 1) # unindo as colunas só pre processadas a cima dentro da variavel x
X = pd.DataFrame(X) #transformando arry em dataframe
X = X.fillna(0) #transformando os NaN do dataframe em 0
Y = base.Survived
```

```python
X_treino, X_teste, Y_treino, Y_teste = train_test_split(X, Y, train_size=2/3, random_state=8)
```

```python
classificador2 = KNeighborsClassifier(n_neighbors=3) #classificador treino_teste 
classificador2.fit(X_treino,Y_treino)

(accuracy_score(Y_teste, classificador2.predict(X_teste)))*100
```

```python
classificador2.predict(X_teste)
```
